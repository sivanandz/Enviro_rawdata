```python
import cdsapi
import os
import time

# ==========================================
# CONFIGURATION
# ==========================================
WORKER_ID = 2  # <--- CHANGE THIS for each worker (1, 2, 3, or 4)

# PASTE THE SPECIFIC WORKER'S CITIES LIST HERE

CITIES = [
    {
        "city": "Whitehorse",
        "country": "Canada",
        "lat": 60.72,
        "lon": -135.05
    },
    {
        "city": "Atlin",
        "country": "Canada",
        "lat": 59.57,
        "lon": -133.71
    },
    {
        "city": "Calgary (Border)",
        "country": "Canada",
        "lat": 51.04,
        "lon": -114.07
    },
    {
        "city": "Iqaluit",
        "country": "Canada",
        "lat": 63.74,
        "lon": -68.51
    },
    {
        "city": "Alert",
        "country": "Canada",
        "lat": 82.5,
        "lon": -62.33
    },
    {
        "city": "Rome",
        "country": "Italy",
        "lat": 41.9,
        "lon": 12.49
    },
    {
        "city": "Rieti",
        "country": "Italy",
        "lat": 42.4,
        "lon": 12.86
    },
    {
        "city": "Milan",
        "country": "Italy",
        "lat": 45.46,
        "lon": 9.18
    },
    {
        "city": "Bologna",
        "country": "Italy",
        "lat": 44.49,
        "lon": 11.34
    },
    {
        "city": "Lampedusa",
        "country": "Italy",
        "lat": 35.51,
        "lon": 12.58
    },
    {
        "city": "Ishigaki",
        "country": "Japan",
        "lat": 24.34,
        "lon": 124.15
    },
    {
        "city": "Chichijima",
        "country": "Japan",
        "lat": 27.09,
        "lon": 142.19
    },
    {
        "city": "Tokyo",
        "country": "Japan",
        "lat": 35.68,
        "lon": 139.69
    },
    {
        "city": "Sapporo",
        "country": "Japan",
        "lat": 43.06,
        "lon": 141.35
    },
    {
        "city": "Kushiro",
        "country": "Japan",
        "lat": 42.98,
        "lon": 144.38
    },
    {
        "city": "Paris",
        "country": "France",
        "lat": 48.85,
        "lon": 2.35
    },
    {
        "city": "Toulouse",
        "country": "France",
        "lat": 43.6,
        "lon": 1.44
    },
    {
        "city": "Nice",
        "country": "France",
        "lat": 43.71,
        "lon": 7.26
    },
    {
        "city": "Briançon",
        "country": "France",
        "lat": 44.89,
        "lon": 6.64
    },
    {
        "city": "Tignes",
        "country": "France",
        "lat": 45.46,
        "lon": 6.9
    },
    {
        "city": "Cayenne (FR Guiana)",
        "country": "France",
        "lat": 4.93,
        "lon": -52.33
    },
    {
        "city": "Villahermosa",
        "country": "Mexico",
        "lat": 17.98,
        "lon": -92.93
    },
    {
        "city": "Cancún",
        "country": "Mexico",
        "lat": 21.16,
        "lon": -86.85
    },
    {
        "city": "Hermosillo",
        "country": "Mexico",
        "lat": 29.07,
        "lon": -110.97
    },
    {
        "city": "Ciudad Juárez",
        "country": "Mexico",
        "lat": 31.69,
        "lon": -106.42
    },
    {
        "city": "Monterrey",
        "country": "Mexico",
        "lat": 25.68,
        "lon": -100.31
    },
    {
        "city": "Pachuca",
        "country": "Mexico",
        "lat": 20.1,
        "lon": -98.75
    },
    {
        "city": "Cuernavaca",
        "country": "Mexico",
        "lat": 18.92,
        "lon": -99.23
    },
    {
        "city": "Mexico City",
        "country": "Mexico",
        "lat": 19.43,
        "lon": -99.13
    },
    {
        "city": "Tecate",
        "country": "Mexico",
        "lat": 32.56,
        "lon": -116.62
    },
    {
        "city": "Matamoros",
        "country": "Mexico",
        "lat": 25.86,
        "lon": -97.5
    },
    {
        "city": "Şanlıurfa",
        "country": "Turkey",
        "lat": 37.16,
        "lon": 38.79
    },
    {
        "city": "Sivas",
        "country": "Turkey",
        "lat": 39.75,
        "lon": 37.01
    },
    {
        "city": "Istanbul",
        "country": "Turkey",
        "lat": 41.0,
        "lon": 28.97
    },
    {
        "city": "Ankara",
        "country": "Turkey",
        "lat": 39.93,
        "lon": 32.85
    },
    {
        "city": "Erzurum",
        "country": "Turkey",
        "lat": 39.9,
        "lon": 41.27
    },
    {
        "city": "Port Harcourt",
        "country": "Nigeria",
        "lat": 4.81,
        "lon": 7.04
    },
    {
        "city": "Lagos",
        "country": "Nigeria",
        "lat": 6.52,
        "lon": 3.37
    },
    {
        "city": "Kano",
        "country": "Nigeria",
        "lat": 12.0,
        "lon": 8.59
    },
    {
        "city": "Almería",
        "country": "Spain",
        "lat": 36.83,
        "lon": -2.46
    },
    {
        "city": "Lorca",
        "country": "Spain",
        "lat": 37.67,
        "lon": -1.7
    },
    {
        "city": "Alicante",
        "country": "Spain",
        "lat": 38.34,
        "lon": -0.48
    },
    {
        "city": "Madrid",
        "country": "Spain",
        "lat": 40.41,
        "lon": -3.7
    },
    {
        "city": "Seville",
        "country": "Spain",
        "lat": 37.38,
        "lon": -5.98
    },
    {
        "city": "Porto (Analog)",
        "country": "Spain",
        "lat": 41.15,
        "lon": -8.62
    },
    {
        "city": "Girona",
        "country": "Spain",
        "lat": 41.97,
        "lon": 2.82
    },
    {
        "city": "Bilbao",
        "country": "Spain",
        "lat": 43.26,
        "lon": -2.93
    },
    {
        "city": "Medan",
        "country": "Indonesia",
        "lat": 3.59,
        "lon": 98.67
    },
    {
        "city": "Jakarta",
        "country": "Indonesia",
        "lat": -6.17,
        "lon": 106.82
    },
    {
        "city": "Denpasar",
        "country": "Indonesia",
        "lat": -8.67,
        "lon": 115.21
    },
    {
        "city": "Da Nang",
        "country": "Vietnam",
        "lat": 16.05,
        "lon": 108.2
    },
    {
        "city": "Ho Chi Minh City",
        "country": "Vietnam",
        "lat": 10.82,
        "lon": 106.63
    },
    {
        "city": "Hanoi",
        "country": "Vietnam",
        "lat": 21.02,
        "lon": 105.83
    },
    {
        "city": "Kyiv",
        "country": "Ukraine",
        "lat": 50.45,
        "lon": 30.52
    },
    {
        "city": "Sevastopol",
        "country": "Ukraine",
        "lat": 44.61,
        "lon": 33.52
    },
    {
        "city": "Bernardo de Irigoyen",
        "country": "Argentina",
        "lat": -26.25,
        "lon": -53.64
    },
    {
        "city": "Chilecito",
        "country": "Argentina",
        "lat": -29.16,
        "lon": -67.49
    },
    {
        "city": "Mendoza",
        "country": "Argentina",
        "lat": -32.88,
        "lon": -68.84
    },
    {
        "city": "La Rioja",
        "country": "Argentina",
        "lat": -29.41,
        "lon": -66.85
    },
    {
        "city": "Viedma",
        "country": "Argentina",
        "lat": -40.81,
        "lon": -62.99
    },
    {
        "city": "Mar del Plata (Border)",
        "country": "Argentina",
        "lat": -38.0,
        "lon": -57.55
    }
]


DATA_DIR = "era5_data_2025"
YEARS = [2025]
MONTHS = list(range(1, 13))

# Ensure data directory exists
os.makedirs(DATA_DIR, exist_ok=True)

def get_city_progress(city, country, data_dir):
    """Check how many files exist for a city and return missing files"""
    existing_files = 0
    missing_files = []
    
    for year in YEARS:
        for month in MONTHS:
            filename = f"{city}_{country}_{year}_{month:02d}.nc"
            filepath = os.path.join(data_dir, filename)
            
            if os.path.exists(filepath):
                existing_files += 1
            else:
                missing_files.append((year, month))
    
    return existing_files, missing_files

def is_city_complete(city, country, data_dir):
    """Check if all files exist for a city"""
    existing, missing = get_city_progress(city, country, data_dir)
    total_expected = len(YEARS) * len(MONTHS)
    return existing == total_expected

def download_era5_data(city, country, lat, lon, year, month):
    """Download ERA5 data for specific city, year, and month"""
    c = cdsapi.Client()
    
    filename = f"{city}_{country}_{year}_{month:02d}.nc"
    filepath = os.path.join(DATA_DIR, filename)
    
    if os.path.exists(filepath):
        print(f"[Worker {WORKER_ID}] File already exists: {filename}")
        return True
    
    try:
        print(f"[Worker {WORKER_ID}] Downloading {filename}...")
        c.retrieve(
            'reanalysis-era5-single-levels',
            {
                'product_type': 'reanalysis',
                'variable': [
                    '2m_temperature',
                    'skin_temperature',
                    '2m_dewpoint_temperature',
                    'total_precipitation',
                    'snowfall',
                    'snow_depth',
                    'snow_density',
                    'downward_uv_radiation_at_the_surface', # UV Radiation
                    'total_cloud_cover',
                    'surface_solar_radiation_downwards',
                    '10m_u_component_of_wind',
                    '10m_v_component_of_wind',
                    'significant_height_of_combined_wind_waves_and_swell',
                    'mean_wave_period',
                    'land_sea_mask', # To filter land data
                ],
                'year': str(year),
                'month': f'{month:02d}',
                'day': [f'{d:02d}' for d in range(1, 32)],
                'time': [f'{h:02d}:00' for h in range(24)],
                'area': [lat, lon, lat, lon], # Nearest Grid Point
                'format': 'netcdf',
            },
            filepath
        )
        print(f"[Worker {WORKER_ID}] Successfully downloaded: {filename}")
        return True
    except Exception as e:
        print(f"[Worker {WORKER_ID}] Error downloading {filename}: {str(e)}")
        # Note: Wave parameters might cause errors or return NaNs for inland cities depending on the grid mask.
        return False

def process_city(city_data):
    """Process all months for a city"""
    city = city_data["city"]
    country = city_data["country"]
    lat = city_data["lat"]
    lon = city_data["lon"]
    
    # Check if city is already complete
    if is_city_complete(city, country, DATA_DIR):
        print(f"[Worker {WORKER_ID}] {city}, {country} already complete. Skipping.")
        return True
    
    # Get progress
    existing, missing = get_city_progress(city, country, DATA_DIR)
    print(f"[Worker {WORKER_ID}] Starting city: {city}, {country}")
    print(f"[Worker {WORKER_ID}] Progress: {existing} files exist, {len(missing)} files to download")
    
    success_count = existing
    
    # Download only missing files
    for year, month in missing:
        if download_era5_data(city, country, lat, lon, year, month):
            success_count += 1
        time.sleep(2)  # Rate limiting
    
    total_expected = len(YEARS) * len(MONTHS)
    print(f"[Worker {WORKER_ID}] Completed {city}: {success_count}/{total_expected} files")
    return success_count == total_expected

def main():
    """Main processing loop"""
    print(f"=== Worker {WORKER_ID} Started ===")
    print(f"Processing {len(CITIES)} cities")
    print(f"Target Year: 2025")
    
    completed_cities = 0
    
    for city_data in CITIES:
        if process_city(city_data):
            completed_cities += 1
    
    print(f"=== Worker {WORKER_ID} Finished ===")
    print(f"Successfully completed: {completed_cities}/{len(CITIES)} cities")

if __name__ == "__main__":
    main()
```
