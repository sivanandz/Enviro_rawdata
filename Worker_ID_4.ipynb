```python
import cdsapi
import os
import time

# ==========================================
# CONFIGURATION
# ==========================================
WORKER_ID = 4  # <--- CHANGE THIS for each worker (1, 2, 3, or 4)

# PASTE THE SPECIFIC WORKER'S CITIES LIST HERE

CITIES = [
    {
        "city": "Guatemala City",
        "country": "Guatemala",
        "lat": 14.63,
        "lon": -90.5
    },
    {
        "city": "Faro",
        "country": "Portugal",
        "lat": 37.01,
        "lon": -7.93
    },
    {
        "city": "Porto",
        "country": "Portugal",
        "lat": 41.15,
        "lon": -8.62
    },
    {
        "city": "Pingtung",
        "country": "Taiwan",
        "lat": 22.67,
        "lon": 120.48
    },
    {
        "city": "Taipei",
        "country": "Taiwan",
        "lat": 25.03,
        "lon": 121.56
    },
    {
        "city": "Kaohsiung",
        "country": "Taiwan",
        "lat": 22.62,
        "lon": 120.3
    },
    {
        "city": "Piraeus",
        "country": "Greece",
        "lat": 37.94,
        "lon": 23.64
    },
    {
        "city": "Thessaloniki",
        "country": "Greece",
        "lat": 40.64,
        "lon": 22.94
    },
    {
        "city": "Patras",
        "country": "Greece",
        "lat": 38.24,
        "lon": 21.73
    },
    {
        "city": "Colombo",
        "country": "Sri_Lanka",
        "lat": 6.92,
        "lon": 79.86
    },
    {
        "city": "Batticaloa",
        "country": "Sri_Lanka",
        "lat": 7.71,
        "lon": 81.7
    },
    {
        "city": "Jaffna",
        "country": "Sri_Lanka",
        "lat": 9.66,
        "lon": 80.01
    },
    {
        "city": "Dar es Salaam",
        "country": "Tanzania",
        "lat": -6.79,
        "lon": 39.2
    },
    {
        "city": "Dodoma",
        "country": "Tanzania",
        "lat": -6.17,
        "lon": 35.75
    },
    {
        "city": "Kuala Lumpur",
        "country": "Malaysia",
        "lat": 3.13,
        "lon": 101.68
    },
    {
        "city": "Langkawi",
        "country": "Malaysia",
        "lat": 6.35,
        "lon": 99.8
    },
    {
        "city": "Cameron Highlands",
        "country": "Malaysia",
        "lat": 4.47,
        "lon": 101.38
    },
    {
        "city": "London",
        "country": "United_Kingdom",
        "lat": 51.5,
        "lon": -0.12
    },
    {
        "city": "Adamstown (Pitcairn)",
        "country": "United_Kingdom",
        "lat": -25.06,
        "lon": -130.1
    },
    {
        "city": "Berlin",
        "country": "Germany",
        "lat": 52.52,
        "lon": 13.4
    },
    {
        "city": "Nuremberg",
        "country": "Germany",
        "lat": 49.45,
        "lon": 11.07
    },
    {
        "city": "Guayaquil",
        "country": "Ecuador",
        "lat": -2.17,
        "lon": -79.92
    },
    {
        "city": "Manta",
        "country": "Ecuador",
        "lat": -0.96,
        "lon": -80.71
    },
    {
        "city": "Portoviejo",
        "country": "Ecuador",
        "lat": -1.05,
        "lon": -80.45
    },
    {
        "city": "Ibarra",
        "country": "Ecuador",
        "lat": 0.35,
        "lon": -78.11
    },
    {
        "city": "Varna",
        "country": "Bulgaria",
        "lat": 43.21,
        "lon": 27.91
    },
    {
        "city": "Sofia",
        "country": "Bulgaria",
        "lat": 42.69,
        "lon": 23.32
    },
    {
        "city": "Biskra",
        "country": "Algeria",
        "lat": 34.85,
        "lon": 5.73
    },
    {
        "city": "Naâma",
        "country": "Algeria",
        "lat": 33.26,
        "lon": -0.31
    },
    {
        "city": "Sétif",
        "country": "Algeria",
        "lat": 36.19,
        "lon": 5.41
    },
    {
        "city": "Algiers",
        "country": "Algeria",
        "lat": 36.75,
        "lon": 3.05
    },
    {
        "city": "Gabès",
        "country": "Tunisia",
        "lat": 33.88,
        "lon": 10.1
    },
    {
        "city": "Sfax",
        "country": "Tunisia",
        "lat": 34.74,
        "lon": 10.76
    },
    {
        "city": "El Kef",
        "country": "Tunisia",
        "lat": 36.17,
        "lon": 8.7
    },
    {
        "city": "Tunis",
        "country": "Tunisia",
        "lat": 36.8,
        "lon": 10.18
    },
    {
        "city": "Aktau",
        "country": "Kazakhstan",
        "lat": 43.64,
        "lon": 51.17
    },
    {
        "city": "Karaganda",
        "country": "Kazakhstan",
        "lat": 49.83,
        "lon": 73.16
    },
    {
        "city": "Almaty",
        "country": "Kazakhstan",
        "lat": 43.22,
        "lon": 76.85
    },
    {
        "city": "Astana",
        "country": "Kazakhstan",
        "lat": 51.16,
        "lon": 71.44
    },
    {
        "city": "Villa Tunari",
        "country": "Bolivia",
        "lat": -16.97,
        "lon": -65.42
    },
    {
        "city": "Trinidad",
        "country": "Bolivia",
        "lat": -14.83,
        "lon": -64.9
    },
    {
        "city": "Tarija",
        "country": "Bolivia",
        "lat": -21.53,
        "lon": -64.73
    },
    {
        "city": "Sucre",
        "country": "Bolivia",
        "lat": -19.03,
        "lon": -65.26
    },
    {
        "city": "El Alto",
        "country": "Bolivia",
        "lat": -16.51,
        "lon": -68.16
    },
    {
        "city": "Montevideo",
        "country": "Uruguay",
        "lat": -34.9,
        "lon": -56.16
    },
    {
        "city": "Bocas del Toro",
        "country": "Panama",
        "lat": 9.34,
        "lon": -82.24
    },
    {
        "city": "David",
        "country": "Panama",
        "lat": 8.43,
        "lon": -82.43
    },
    {
        "city": "Warsaw",
        "country": "Poland",
        "lat": 52.22,
        "lon": 21.01
    },
    {
        "city": "Dublin",
        "country": "Ireland",
        "lat": 53.34,
        "lon": -6.26
    },
    {
        "city": "Dalanzadgad",
        "country": "Mongolia",
        "lat": 43.57,
        "lon": 104.42
    },
    {
        "city": "Ulaanbaatar",
        "country": "Mongolia",
        "lat": 47.91,
        "lon": 106.91
    },
    {
        "city": "Bulgan",
        "country": "Mongolia",
        "lat": 48.81,
        "lon": 103.53
    },
    {
        "city": "Tsetserleg",
        "country": "Mongolia",
        "lat": 47.47,
        "lon": 101.45
    },
    {
        "city": "Belgrade",
        "country": "Serbia",
        "lat": 44.78,
        "lon": 20.44
    },
    {
        "city": "Vienna",
        "country": "Austria",
        "lat": 48.2,
        "lon": 16.37
    },
    {
        "city": "Salzburg",
        "country": "Austria",
        "lat": 47.8,
        "lon": 13.04
    },
    {
        "city": "Obergurgl",
        "country": "Austria",
        "lat": 46.87,
        "lon": 11.02
    },
    {
        "city": "Split (Border)",
        "country": "Croatia",
        "lat": 43.5,
        "lon": 16.44
    },
    {
        "city": "Split",
        "country": "Croatia",
        "lat": 43.5,
        "lon": 16.44
    },
    {
        "city": "Zagreb",
        "country": "Croatia",
        "lat": 45.81,
        "lon": 15.97
    }
]

DATA_DIR = "era5_data_2025"
YEARS = [2025]
MONTHS = list(range(1, 13))

# Ensure data directory exists
os.makedirs(DATA_DIR, exist_ok=True)

def get_city_progress(city, country, data_dir):
    """Check how many files exist for a city and return missing files"""
    existing_files = 0
    missing_files = []
    
    for year in YEARS:
        for month in MONTHS:
            filename = f"{city}_{country}_{year}_{month:02d}.nc"
            filepath = os.path.join(data_dir, filename)
            
            if os.path.exists(filepath):
                existing_files += 1
            else:
                missing_files.append((year, month))
    
    return existing_files, missing_files

def is_city_complete(city, country, data_dir):
    """Check if all files exist for a city"""
    existing, missing = get_city_progress(city, country, data_dir)
    total_expected = len(YEARS) * len(MONTHS)
    return existing == total_expected

def download_era5_data(city, country, lat, lon, year, month):
    """Download ERA5 data for specific city, year, and month"""
    c = cdsapi.Client()
    
    filename = f"{city}_{country}_{year}_{month:02d}.nc"
    filepath = os.path.join(DATA_DIR, filename)
    
    if os.path.exists(filepath):
        print(f"[Worker {WORKER_ID}] File already exists: {filename}")
        return True
    
    try:
        print(f"[Worker {WORKER_ID}] Downloading {filename}...")
        c.retrieve(
            'reanalysis-era5-single-levels',
            {
                'product_type': 'reanalysis',
                'variable': [
                    '2m_temperature',
                    'skin_temperature',
                    '2m_dewpoint_temperature',
                    'total_precipitation',
                    'snowfall',
                    'snow_depth',
                    'snow_density',
                    'downward_uv_radiation_at_the_surface', # UV Radiation
                    'total_cloud_cover',
                    'surface_solar_radiation_downwards',
                    '10m_u_component_of_wind',
                    '10m_v_component_of_wind',
                    'significant_height_of_combined_wind_waves_and_swell',
                    'mean_wave_period',
                    'land_sea_mask', # To filter land data
                ],
                'year': str(year),
                'month': f'{month:02d}',
                'day': [f'{d:02d}' for d in range(1, 32)],
                'time': [f'{h:02d}:00' for h in range(24)],
                'area': [lat, lon, lat, lon], # Nearest Grid Point
                'format': 'netcdf',
            },
            filepath
        )
        print(f"[Worker {WORKER_ID}] Successfully downloaded: {filename}")
        return True
    except Exception as e:
        print(f"[Worker {WORKER_ID}] Error downloading {filename}: {str(e)}")
        # Note: Wave parameters might cause errors or return NaNs for inland cities depending on the grid mask.
        return False

def process_city(city_data):
    """Process all months for a city"""
    city = city_data["city"]
    country = city_data["country"]
    lat = city_data["lat"]
    lon = city_data["lon"]
    
    # Check if city is already complete
    if is_city_complete(city, country, DATA_DIR):
        print(f"[Worker {WORKER_ID}] {city}, {country} already complete. Skipping.")
        return True
    
    # Get progress
    existing, missing = get_city_progress(city, country, DATA_DIR)
    print(f"[Worker {WORKER_ID}] Starting city: {city}, {country}")
    print(f"[Worker {WORKER_ID}] Progress: {existing} files exist, {len(missing)} files to download")
    
    success_count = existing
    
    # Download only missing files
    for year, month in missing:
        if download_era5_data(city, country, lat, lon, year, month):
            success_count += 1
        time.sleep(2)  # Rate limiting
    
    total_expected = len(YEARS) * len(MONTHS)
    print(f"[Worker {WORKER_ID}] Completed {city}: {success_count}/{total_expected} files")
    return success_count == total_expected

def main():
    """Main processing loop"""
    print(f"=== Worker {WORKER_ID} Started ===")
    print(f"Processing {len(CITIES)} cities")
    print(f"Target Year: 2025")
    
    completed_cities = 0
    
    for city_data in CITIES:
        if process_city(city_data):
            completed_cities += 1
    
    print(f"=== Worker {WORKER_ID} Finished ===")
    print(f"Successfully completed: {completed_cities}/{len(CITIES)} cities")

if __name__ == "__main__":
    main()
```
